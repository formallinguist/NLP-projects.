WORD EMBEDDING

1. CONVERTING TEXT TO VECTOR
2. COUNTING THE OCCURANCE OF EVERY WORD IN VOCAB(BOW METHOD)
3. FIND FREQUENCY OF THE WORD AND SEE HOW IMPORTANT THE WORD IS IN THE DOCUMENT (TF-IDF).
4.These are classical methods, Bow and Tf/idf cannot understand the word in the context.




```python
import nltk
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from nltk.tokenize import sent_tokenize
```

```python
text = '''An apple is an edible fruit produced by an apple tree (Malus domestica).
 Apple trees are cultivated worldwide and are the most widely grown species in the genus Malus. 
 The tree originated in Central Asia, where its wild ancestor, Malus sieversii, is still found today.
 Apples have been grown for thousands of years in Asia and Europe and were brought to North America by European colonists.
 Apples have religious and mythological significance in many cultures, including Norse, Greek, and European Christian tradition.'''
```

```python
Text = sent_tokenize(text)
```

```python
Text
```

```python
def preprocess(document):
  document = document.lower()
  words = word_tokenize(document)
  words = [word for word in words if word not in stopwords.words('english')]
  document = " ".join(words)
  return document
```

```python
ptext = [preprocess(document) for document in Text]
```

```python
ptext
```

CONVERSION OF TEXT TO NUMBERS IS KNOWN AS VECTORS.

```python
Vectorizer = CountVectorizer()
```

```python
Bow = Vectorizer.fit_transform(ptext)
```

Bag of words model.

```python
print(Bow.toarray())
```

```python
len(Vectorizer.get_feature_names())
```

```python
import pandas as pd
```

```python
pd.DataFrame(Bow.toarray(),columns=Vectorizer.get_feature_names())
```

TF/IDF.
1. Term Frequency.
2. Inverse Document Frequency.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
ptext = [preprocess(document) for document in Text]
```

```python
Tf_Vec= TfidfVectorizer()
```

```python
tfidf= Tf_Vec.fit_transform(ptext)
```

```python
pd.DataFrame(tfidf.toarray(),columns=Tf_Vec.get_feature_names())
```

```python
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer

stemmer = PorterStemmer()
word_lemmatizer = WordNetLemmatizer()
```

APPLYING STEMMING AND LEMMATIZATION TO THE CORPUS.

```python
def preprocess(document):
  document = document.lower()
  words = word_tokenize(document)
  words = [word for word in words if word not in stopwords.words('english')]
  words = [word_lemmatizer.lemmatize(word, pos='v') for word in words]
  #words = [stemmer.stem(word) for word in words]
  document = " ".join(words)
  return document
```

```python
nltk.download('wordnet')
```

```python
ptext = [preprocess(document) for document in Text]
Tf_Vec= TfidfVectorizer()
tfidf= Tf_Vec.fit_transform(ptext)
pd.DataFrame(tfidf.toarray(),columns=Tf_Vec.get_feature_names())

```
