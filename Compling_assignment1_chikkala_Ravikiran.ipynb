#Zipf's law for English texts, Jungle book and KJV.

```python

#For English texts I performed some preprocesing steps like lowering of texts,removing the numericals and special characters.

# Here, I Created a function where it processes the both english text files
def process_english_text_files(filename):
  # Here, I open the files with open and encoding utf-8.
    with open(filename, "r", encoding="utf-8") as file:
       #Here I am reading and lowering the text
        text = file.read().lower()
        #I used regex to get only words from the text
        words = re.findall(r'\b[a-zA-Z]+\b', text)
     #Here I am printing file name
    print(f"\nAnalysis for {filename}:")

    # Here, we get to see the length of words of the text
    print(f"Total words: {len(words)}")

    # Here, we are uisng set function to count vocabulary
    unique_words = set(words)
    print(f"Unique words: {len(unique_words)}")

    # Here, I calculated the word frequencies using nltk Frequency distribution function
    frequency_dist = nltk.FreqDist(words)

    # Here, I see the top ten frequently occurring words in both texts
    print("\nTop ten frequently occurring words:")
    for word, frequency in frequency_dist.most_common(10):
        print(f"{word}: {frequency}")

    # Here, words are sorted based on highest frequency by uisng the sorted function
    high_frequent_words = sorted(frequency_dist.values(), reverse=True)

    #Here, I Plotted frequencies in linear scale with X-axis having 'position in the frequency list' with Y-axis as 'frequency'.

    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.plot(high_frequent_words)
    plt.xlabel('Position in Frequency List')
    plt.ylabel('Frequency')
    plt.title('Frequency Curve (Linear Axes)')

  # Here, I Plotted frequencies in log-log scale with X-axis having 'position in the frequency list' with Y-axis as 'frequency'.
    plt.figure(figsize=(10, 5))
    plt.subplot(122)
    plt.loglog(high_frequent_words)
    plt.xlabel('Position in Frequency List (Log Scale)')
    plt.ylabel('Frequency (Log Scale)')
    plt.title('Frequency Curve (Log-Log Axes)')
  # Here, we are calling 'plt.show()' to display the plot.
    plt.show()
```

```python
#Here, I am calling the function "process_english_text_files" on both the english texts.
process_english_text_files(jungle_book)
process_english_text_files(kjv_bible)
```

```python
nltk.download('punkt')
from nltk.probability import (FreqDist, ConditionalFreqDist, ConditionalProbDist, MLEProbDist, SimpleGoodTuringProbDist)
from nltk.util import ngrams
```

```python
''' Here, this function uses the Maximum Likelihood Estimation (MLE) method
to create a probability distribution from a frequency distribution.'''

def ml_estimator(freqdist):
    return MLEProbDist(freqdist)

'''Here,this function uses the good_turing frequency estimation
method to create a probability distribution from a frequency distribution '''

def goodturing_estimator(freqdist):
    return SimpleGoodTuringProbDist(freqdist)
```

```python
'''Here, I used the python class BsicNgram for training n-gram models from the
assignment that we were asked to use from the assignment.'''

class BasicNgram(ConditionalProbDist):
    """
    Define and train an Ngram Model over the corpus represented by the list words.
    Given an BasicNgram instance ngram and a (n-1)-gram context (i.e., a tuple of n-1 strings),
    a call to ngram[context] returns a nltk.probability.ProbDistI object representing the Probability distribution P(.|context) over possible values for the next word.
    Be aware that context has to be a tuple, even if context is a unigram (see example below)

    >>> corpus=['a','b','b','a']
    >>> bigram=BasicNgram(2,corpus)
    >>> bigram.contexts()
    [('<$>',), ('a',), ('b',)]
    >>> p_b=bigram[('b',)] #not bigram['b']!!!
    >>> p_b.prob('a')
    0.5
    >>> p_b.prob('b')
    0.5

    :param n: the dimension of the n-grams (i.e. the size of the context+1).
    :type n: int
    :param corpus:
    :type corpus: list(Str)

    other parameters are optional and may be omitted. They define whether to add artificial symbols before or after the word list,
    and whether to use another estimation methods than maximum likelihood.
    """
    def __init__(self, n, words, start_symbol="<$>", end_symbol="</$>", pad_left=True, pad_right=False, estimator=ml_estimator):
        assert (n > 0)
        self._n=n
        self._words=words
        self._counter=ConditionalFreqDist()
        self._start_symbol=start_symbol
        self._end_symbol=end_symbol
        self._pad_left=pad_left
        self._pad_right=pad_right
        self._train()
        super().__init__(self._counter, estimator)


    def _train(self):
        _ngrams=self.generate_ngrams()
        for ngram in _ngrams:
            context=ngram[0:-1]
            outcome=ngram[-1]
            self._counter[context][outcome]+=1

    """
    returns an iterable over the ngrams of the word corpus
    """
    def generate_ngrams(self):
        return ngrams(self._words, self._n, pad_left=self._pad_left, pad_right=self._pad_right,
                      left_pad_symbol=self._start_symbol,
                      right_pad_symbol=self._end_symbol)


    """
    Return the list of contexts
    """
    def contexts(self):
        return list(self.conditions())
```

#Problem2.Producing the text samples of 100 words each.

```python
'''Here, this function generates a sequence of words based on n-gram model and an initial context.
 It does this by repeatedly generating a new word based on the current context and
 updating the context for the next iteration. The generated sequence of words is returned as a single string.'''


import random
# Here, this function takes three parameters ngram, context and num_words.
def generate_text(ngram, context, num_words):
  # here, it has words containing the words in intial context.
    text = list(context)
    #for every iteration iteration a new word is generated and added to text
    for _ in range(num_words):
      # it gives the probability distribution for the current context from ngram model.
        prob_distribution = ngram[context]
        #it generates a new word based on probability distribution.
        next_word = prob_distribution.generate()
        # newly generated word is added to the end of text
        text.append(next_word)
        #it updates the context by assing new word.
        context = tuple((list(context) + [next_word])[-ngram._n+1:])
        # output is the words joined into the single string.
    return ' '.join(text)
```

```python
# Here we see the values of N which are seen in the order of N-garm model.
n_values = [2, 3, 4]

# Here for loop generates text as it iterates over each value of N
for n in n_values:
  # Here, it prints the current value of n
    print(f"\nGenerating text for N={n}:\n")
    #it creates an ngram model with current value of n on the words of jungle book and maximum likelihood estimator.
    ngram = BasicNgram(n=n, words=jungle_book_words, estimator=ml_estimator)
    for _ in range(5):  # Here it generates 5 text samples for each N
        start_context = random.choice(list(ngram.conditions()))  # it randomly chooses a start context from n-gram model.
        print(generate_text(ngram, start_context, 100)) #In the end it generates a text sample of 100 words.
```

#Question3. Statistical dependence.

```python
import math
from collections import defaultdict

# Here, I loaded the corpus from junglebook.txt
with open(jungle_book, "r", encoding="utf-8") as file:
    corpus = file.read().lower()  # Read the corpus and converted it to lowercase
```

```python
#Here, I used split() function to tokenize the corpus.
words = corpus.split()
```

```python
# Here, I counted the frequency of each word using the defaultdict.
word_freq = defaultdict(int)
for word in words:
    word_freq[word] += 1
```

```python
# Here, I removed the words that occur less than 10 times in the corpus
words = [word for word in words if word_freq[word] >= 10]
```

```python
# Here, I recalculated the frequency of each word after removing the words that occur less than 10 times.
word_freq = defaultdict(int)
for word in words:
    word_freq[word] += 1
```

```python

#The function takes four arguments  word1 and word2, (word_freq) and (words).

def calculate_pmi(word1, word2, word_freq, words):

 # Here, it counts how many times word1 and word2 appear together.
    count_w1_w2 = sum(1 for i in range(len(words) - 1) if words[i] == word1 and words[i + 1] == word2)
    # Here it calculates the total occurrences of word1 and word2 seperately in a list.
    count_w1 = word_freq[word1]
    count_w2 = word_freq[word2]
    # Finally it calculates the total number of words in the corpus.
    num_words = len(words)
    # Here, we check for cases where any count is zero inorder to avoid division by zero
    if count_w1 == 0 or count_w2 == 0 or count_w1_w2 == 0:
        return 0.0
    # Used this formula to calculate Pointwise Mutual Information.
    pmi = math.log2((count_w1_w2 * num_words) / (count_w1 * count_w2))
    #Finally it returns pmi value
    return pmi


# Here it calculated the PMI for all word pairs
word_pairs = set()
pmi_values = {}

# It loops over the pair of words in the 'words' list
for i in range(len(words) - 1):
  # Here, I extracted two words from the list based on current and next indexes
    word1, word2 = words[i], words[i + 1]
    # Here, i added the tuple (word1, word2) to the set
    word_pairs.add((word1, word2))
    #Here, we calculate pmi for current word pair
    pmi = calculate_pmi(word1, word2, word_freq, words)
    #Here, it stores the current word pair in a dictionary
    pmi_values[(word1, word2)] = pmi
```

```python
# Here, I sorted the word pairs by PMI in descending order.
sorted_pmi = sorted(pmi_values.items(), key=lambda item: item[1], reverse=True)

# Here, I get the top 20 word pairs with the highest PMI values
top_20_pmi = sorted_pmi[:20]

# Here I get the 20 word pairs with the lowest PMI values
bottom_20_pmi = sorted_pmi[-20:]

# Here is the out put of pmi values and word pairs for highest and lowest values.
print("Top 20 word pairs with highest PMI values:")
for pair, pmi in top_20_pmi:
    print(pair, pmi)

print("\nBottom 20 word pairs with lowest PMI values:")
for pair, pmi in bottom_20_pmi:
    print(pair, pmi)
```
